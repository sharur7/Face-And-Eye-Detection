```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Creating a Classifier
face_classifier = cv2.CascadeClassifier('Haarcascade/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascade/haarcascade_eye.xml')
```

```python
#Open Image
image = cv2.imread('Images/image4.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Detect Face
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

```python
if faces is ():
    print('No Faces found')
```

```python
#Its a kind of Magic
def detect_faces_eyes(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img, (x,y), (x+w, y+h), (255,0,0), 7)
        
    eyes_rects = eye_classifier.detectMultiScale(fix_img)
    
    for(ix,iy,iw,ih) in eyes_rects:
        cv2.rectangle(fix_img, (ix,iy), (ix+iw, iy+ih), (0,0,255), 5)    
        
        return fix_img
    
```

```python
result = detect_faces_eyes(fix_img)
plt.imshow(result)
```
