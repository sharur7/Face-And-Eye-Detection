```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Creating a Classifier
eye_classifier = cv2.CascadeClassifier('Haarcascade/haarcascade_eye.xml')
```

```python
#Open Image
image = cv2.imread('Images/image1.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Detect eyes
eyes = eye_classifier.detectMultiScale(image, 1.3, 5)
```

```python
if eyes is ():
    print('No eyes found')
```

```python
#Its a kind of Magic
def detect_eyes(fix_img):
    eyes_rects = eye_classifier.detectMultiScale(fix_img)
    
    for(x,y,w,h) in eyes_rects:
        cv2.rectangle(fix_img, (x,y), (x+w, y+h), (255,255,255), 10)
        
        return fix_img
```

```python
result = detect_eyes(fix_img)
plt.imshow(result)
```
